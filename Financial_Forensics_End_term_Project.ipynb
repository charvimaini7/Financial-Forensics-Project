Approach 1
filter stocks whose price has reduced from T1 to T2 then conduct a ratio analysis (in one the sheets we already have data regarding the ratios so you won't need to calculate from scratch)
For each ratio you'll need to identify if it's beneficial for the ratio to increase or decrease

The final list would be stocks whose prices have decreased but the ratios have improved and hence there's a buying opportunity

Ratios identified to be beneficial to increase:

* Dividend Yield
* Return on Assets
* Return on Equity

Ratios identified to be beneficial to decrease:
* Price to Earning (PE)
* Price to Book Value
* Debt to Equity

```python
# Load the two CSV files
price_file = "price.csv"  # Replace with the path to your first file
price_T2_file = "price_T2.csv"  # Replace with the path to your second file

price_df = pd.read_csv(price_file)
price_T2_df = pd.read_csv(price_T2_file)

# Merge the two DataFrames on the stock identifier (assuming the column is called 'Stock')
merged_data = pd.merge(price_df, price_T2_df, on='Name', suffixes=('_T1', '_T2'))

# Filter stocks where the price has reduced
price_reduced_stocks = merged_data[merged_data['Current Price_T2'] < merged_data['Current Price_T1']]

# Save the filtered stocks to a new CSV file
output_file = "stocks_with_reduced_price.csv"
price_reduced_stocks.to_csv(output_file, index=False)

# Print the result and show the filtered stocks
print(f"Filtered {len(price_reduced_stocks)} stocks where the price has reduced.")
display(price_reduced_stocks)
```

```python
# Load the two CSV files
ratio_file = "ratios_1.csv"  # Replace with the path to your first file
ratio_2_file = "ratios_1_T2.csv"  # Replace with the path to your second file

ratio_df = pd.read_csv(ratio_file)
ratio_2_df = pd.read_csv(ratio_2_file)

# Merge the two DataFrames on a common identifier, e.g., 'Stock'
merged_data = pd.merge(ratio_df, ratio_2_df, on='Name', suffixes=('_T1', '_T2'))

# Filter rows where the specified columns have increased
increased_ratios = merged_data[
    (merged_data['Dividend yield_T2'] > merged_data['Dividend yield_T1']) &
    (merged_data['Return on assets_T2'] > merged_data['Return on assets_T1']) &
    (merged_data['Return on equity_T2'] > merged_data['Return on equity_T1'])
]

# Save the filtered stocks to a new CSV file
output_file = "increased_ratios.csv"
increased_ratios.to_csv(output_file, index=False)

# Print the result and display the filtered rows
print(f"Filtered {len(increased_ratios)} stocks where ratios have increased.")
display(increased_ratios)
```

```python
# Load the two CSV files
ratio_file = "ratios_1.csv"  # Replace with the path to your first file
ratio_2_file = "ratios_1_T2.csv"  # Replace with the path to your second file

ratio_df = pd.read_csv(ratio_file)
ratio_2_df = pd.read_csv(ratio_2_file)

# Merge the two DataFrames on a common identifier, e.g., 'Stock'
merged_data = pd.merge(ratio_df, ratio_2_df, on='Name', suffixes=('_T1', '_T2'))

# Filter rows where:
# 1. Price has reduced
# 2. PE, PBV, and Debt to Equity have decreased
filtered_data = merged_data[
    (merged_data['Price to Earning_T2'] < merged_data['Price to Earning_T1']) &
    (merged_data['Price to book value_T2'] < merged_data['Price to book value_T1']) &
    (merged_data['Debt to equity_T2'] < merged_data['Debt to equity_T1'])
]

# Save the filtered stocks to a new CSV file
output_file = "decreased_ratios.csv"
filtered_data.to_csv(output_file, index=False)

# Print the result and display the filtered rows
print(f"Filtered {len(filtered_data)} stocks where ratios improved.")
display(filtered_data)
```

```python
import pandas as pd

# Load the necessary CSV files
price_file = "price.csv"  # File with price data for T1
price_T2_file = "price_T2.csv"  # File with price data for T2
ratio_file = "ratios_1.csv"  # File with ratio data for T1
ratio_T2_file = "ratios_1_T2.csv"  # File with ratio data for T2

# Load the data
price_df = pd.read_csv(price_file)
price_T2_df = pd.read_csv(price_T2_file)
ratio_df = pd.read_csv(ratio_file)
ratio_T2_df = pd.read_csv(ratio_T2_file)

# Merge price data to compare the current price from T1 and T2
merged_price_data = pd.merge(price_df, price_T2_df, on='Name', suffixes=('_T1', '_T2'))

# Filter for stocks where the price has decreased
reduced_price_stocks = merged_price_data[merged_price_data['Current Price_T1'] > merged_price_data['Current Price_T2']]

# Merge ratio data
merged_ratio_data = pd.merge(ratio_df, ratio_T2_df, on='Name', suffixes=('_T1', '_T2'))

# Filter for stocks where ratios have improved
improved_ratios = merged_ratio_data[
    (merged_ratio_data['Dividend yield_T2'] > merged_ratio_data['Dividend yield_T1']) &
    (merged_ratio_data['Return on assets_T2'] > merged_ratio_data['Return on assets_T1']) &
    (merged_ratio_data['Return on equity_T2'] > merged_ratio_data['Return on equity_T1']) &
    (merged_ratio_data['Price to Earning_T2'] < merged_ratio_data['Price to Earning_T1']) &
    (merged_ratio_data['Price to book value_T2'] < merged_ratio_data['Price to book value_T1']) &
    (merged_ratio_data['Debt to equity_T2'] < merged_ratio_data['Debt to equity_T1'])
]

# Merge the reduced price stocks and improved ratios to find stocks that meet both criteria
final_stocks = pd.merge(reduced_price_stocks, improved_ratios, on='Name')

# Save the result to a new CSV
final_stocks.to_csv("stocks_with_reduced_price_and_improved_ratios.csv", index=False)

# Print the result
print(f"Filtered {len(final_stocks)} stocks that have reduced prices and improved ratios.")
```
